```python
def extract_single_place(url):
    results=requests.get(url)
    soup=bs4.BeautifulSoup(results.text)
    title=soup.find_all('h1',{'class':"DDPage__header-title"})[0].contents[0]
    location=soup.find_all('div',{'class':"DDPage__header-place-location"})[0].contents[0]
    if type(location) ==bs4.element.NavigableString:
        location=location.text
    else:
        location=location.contents[0].text
    short_descriptions=soup.find_all('h3',{'class':"DDPage__header-dek"})[0].contents[0]
    long_description=[x.text for x in soup.find_all('div',{'id':'place-body'})]
    return  title,location,short_descriptions,long_description
```

```python
extract_single_place(url)
```
